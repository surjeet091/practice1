```python
sub_pages = []
for i in pages_list:
    content = requests.get(i,headers=header_list[1])
    soup = bs(content.text,'html.parser')
    for link in soup.findAll('a',class_="_15_ydu6b"):
        req = "https://www.tripadvisor.in/"+str(link.get('href'))
        sub_pages.append(req)
    break
```

```python
import pandas as pd
import random 
import time
data = pd.DataFrame(columns=['Name','OverAll_rating','Num_of_reviews','Range_Cuisine','Address','Contact_No','Rank','Details'])
r=0
for i in sub_pages[::-1]:
    content = requests.get(i)
    print(content)
    time.sleep(int(random.random()*5))
    soup = bs(content.text,'html.parser')
    try:
        Name = soup.find('h1',class_="_3a1XQ88S").text
    except:
        Name = "N/A"
    try:
        OverAll_rating = soup.find('span',class_="r2Cf69qf").text
    except:
        OverAll_rating = "N/A"
    try:
        str_num_of_reviews = soup.find('span',class_="_3Wub8auF").text
        Num_of_reviews = str_num_of_reviews.split(" ")[0] 
    except:
        Num_of_reviews = "N/A"
        range_n_cuisine = []
    try:
        for i in soup.findAll('a',attrs={'class':"_2mn01bsa"}):
            range_n_cuisine.append(i.text)
            Range_Cuisine = range_n_cuisine
    except:
        Range_Cuisine = "N/A"
    try:
        Address = soup.find('span',class_="_2saB_OSe").text
    except:
        Address = "N/A"
    Rev_n_Ph = []
    try:
        j = 1
        for i in soup.findAll('a',class_="_3S6pHEQs"):
            Rev_n_Ph.append(i.text)
            j = j+1
        if i==3:
            break
        Contact_No = Rev_n_Ph[1]
    except:
        Contact_No = "N/A"
    comp =[]
    for i in soup.findAll('a',class_="_15QfMZ2L"):
        comp.append(i.text)
    try:
        Rank = comp[0].split(" ")[0][1:]
    except:
        Rank = "N/A"
    Details = {}
    keys = []
    values = []
    for i in soup.findAll('div',class_="_14zKtJkz"):
        keys.append(i.text)
    for i in soup.findAll('div',class_="_1XLfiSsv"):
        values.append(i.text)
    Details = dict(zip(keys,values))
    row = [Name,OverAll_rating,Num_of_reviews,Range_Cuisine,Address,Contact_No,Rank,Details]
    print(Details)
    data.loc[r] = row
    r = r+1
    print(r,end='  ')
```
